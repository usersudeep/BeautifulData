```python
from bs4 import BeautifulSoup
import requests
import smtplib
from datetime import datetime
import time
```

```python
url="https://hianime.to/home"
headers={"Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8"}
bc=requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(bc.content, 'html.parser')
soup=BeautifulSoup(soup.prettify(),'html.parser')
print(soup)
```

```python
c1=soup.find(id='slider')
print(c1)
```

```python
c2=c1.find_all(class_="desi-head-title dynamic-name")
c4=c1.find_all(class_="desi-description")
c5=[]
```

```python
l = []
print("Spotlight Animes at present:\n")
for i in c2:
    k = []
    k.append("Title:" +i.get_text().strip())
    l.append(k)
for j, i4 in enumerate(c4):
    l[j].append("Description:"+i4.get_text().strip())
for index, row in enumerate(l):
    print(index + 1 )
    for data in row:
        print(data.strip())
    print("-" * 150)
```
